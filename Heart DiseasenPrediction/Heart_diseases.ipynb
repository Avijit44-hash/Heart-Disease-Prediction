```python
import pandas as pd
dataset=pd.read_csv("heart.csv")
dataset.head()
```

```python
inputs=dataset.drop("target",axis="columns")
target_attribute=dataset["target"]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_age=LabelEncoder()
le_sex=LabelEncoder()
le_cp=LabelEncoder()
le_trestbps=LabelEncoder()
le_chol=LabelEncoder()
le_fbs=LabelEncoder()
le_restecg=LabelEncoder()
le_thalach=LabelEncoder()
le_exang=LabelEncoder()
le_oldpeak=LabelEncoder()
le_slope=LabelEncoder()
le_ca=LabelEncoder()
le_thal=LabelEncoder()
```

```python
inputs["age_n"]=le_age.fit_transform(inputs["age"])
inputs["sex_n"]=le_age.fit_transform(inputs["sex"])
inputs["cp_n"]=le_age.fit_transform(inputs["cp"])
inputs["trestbps_n"]=le_age.fit_transform(inputs["trestbps"])
inputs["chol_n"]=le_age.fit_transform(inputs["chol"])
inputs["fbs_n"]=le_age.fit_transform(inputs["fbs"])
inputs["restecg_n"]=le_age.fit_transform(inputs["restecg"])
inputs["thalach_n"]=le_age.fit_transform(inputs["thalach"])
inputs["exang_n"]=le_age.fit_transform(inputs["exang"])
inputs["oldpeak_n"]=le_age.fit_transform(inputs["oldpeak"])
inputs["slope_n"]=le_age.fit_transform(inputs["slope"])
inputs["ca_n"]=le_age.fit_transform(inputs["ca"])
inputs["thal_n"]=le_age.fit_transform(inputs["thal"])

inputs.head()
```

```python
inputs_n=inputs.drop(["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal"],axis="columns")
inputs_n
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(inputs_n,target_attribute)
```

```python
Accuracy=dtree.score(inputs_n,target_attribute)*100
print('\n The accuracy of Decission Tree model :', "%.2f" %Accuracy,"%")
```
